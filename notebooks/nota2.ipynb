```python
#read TRAF with vaex

import vaex

traf = vaex.open('TRAFICO/traf.hdf5')

traf
```

```python
#filter accis by id equal to diff

acci = acci[['fecha', 'hora', 'id']]
acci['accidente'] = 1

acci['id'] = acci['id'].astype('string')

#get rid of the last 2 characters from id

acci['id'] = acci['id'].str.slice(0, -2)

#add 0 to the left of hora if length is 4

for i in range(len(acci)):
    if len(acci['hora'][i]) == 4:
        acci['hora'][i] = '0' + acci['hora'][i]

#acci to vaex

acci_vaex = vaex.from_pandas(acci)

acci_vaex['unir'] = acci_vaex['fecha'] + ' ' + acci_vaex['hora'] + ':00' + ' ' + acci_vaex['id']
acci_vaex = acci_vaex.drop(['fecha','hora','id'])

#group by unir and sum

acci_vaex = acci_vaex.groupby('unir', agg={'accidente':'sum'})

acci_vaex
```

```python
#make unir in traf from fecha_full and idelem

traf['unir'] = traf['fecha'] + ' ' + traf['idelem'].astype('string')

traf
```

```python
#difference between acci and traf in unir

acci_unir = acci_vaex['unir'].values

traf_unir = traf['unir'].values
```

```python
diff = np.setdiff1d(acci_unir, traf_unir)
```
