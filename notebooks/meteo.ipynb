```python
#read geo_loc

geo_loc = pd.read_csv('../../TRAFICO/geo_loc.csv')

geo_loc
```

```python
#drop unnamed column

geo_loc = geo_loc.drop(['Unnamed: 0'], axis=1)
```

```python
#search for closest station to each id

from geopy.distance import geodesic

for i in range(len(geo_loc)):

    dist = []

    for j in range(len(estaciones)):

        dist.append(geodesic((geo_loc.iloc[i]['latitud'], geo_loc.iloc[i]['longitud']), (estaciones.iloc[j]['latitud'], estaciones.iloc[j]['longitud'])).km)

    geo_loc.loc[i, 'station'] = estaciones.iloc[dist.index(min(dist))]['indicativo']

geo_loc
```
