```python
# divide the dataframe into two dataframes based on the null values in the column coordenada_x_utmç
accidentes_null = accidentes[accidentes['coordenada_x_utm'].isnull()]
accidentes_not_null = accidentes[accidentes['coordenada_x_utm'].notnull()]
```

```python
#delete the column 'Unnamed: 0'
geo_loc = geo_loc.drop('Unnamed: 0', axis=1)
```

```python
# from accidents, round to nearest integer the columns 'coordenada_x_utm' and 'coordenada_y_utm'
accidentes_not_null['coordenada_x_utm'] = accidentes_not_null['coordenada_x_utm'].str.split(',', expand=True)[0]
accidentes_not_null['coordenada_y_utm'] = accidentes_not_null['coordenada_y_utm'].str.split(',', expand=True)[0]
```

```python
# Transform the rows with '#¡VALOR!' to NaN
accidentes_not_null['coordenada_x_utm'] = accidentes_not_null['coordenada_x_utm'].replace('#¡VALOR!', np.nan)
accidentes_not_null['coordenada_y_utm'] = accidentes_not_null['coordenada_y_utm'].replace('#¡VALOR!', np.nan)
```

```python
#concatenate the two dataframes
accidentes = pd.concat([accidentes_not_null, accidentes_null])

accidentes
```

```python
# separate the dataframe into two dataframes based on the null values in the column 'coordenada_x_utm'
accidentes_null = accidentes[accidentes['coordenada_x_utm'].isnull()]
accidentes_not_null = accidentes[accidentes['coordenada_x_utm'].notnull()]
```

```python
# transform the columns 'coordenada_x_utm' and 'coordenada_y_utm' to integer
accidentes_not_null['coordenada_x_utm'] = accidentes_not_null['coordenada_x_utm'].astype(int)
accidentes_not_null['coordenada_y_utm'] = accidentes_not_null['coordenada_y_utm'].astype(int)
```

```python
#Find the nearest point in the dataframe geo_loc to the point in the dataframe accidentes
def find_nearest_point(row):

    # get the point in the dataframe accidentes
    point = (row['coordenada_x_utm'], row['coordenada_y_utm'])

    # find the nearest point in the dataframe geo_loc
    nearest_point = geo_loc.iloc[(geo_loc[['utm_x', 'utm_y']].sub(point).pow(2).sum(1).pow(0.5)).idxmin()]

    # return the nearest point
    return nearest_point
```

```python
accidentes_not_null
```

```python
# apply the function to the dataframe accidentes
accidentes_not_null_points = accidentes_not_null.apply(find_nearest_point, axis=1)
```

```python
accidentes_not_null_points
```

```python
# join the two dataframes on the index
accidentes_n = accidentes_not_null_points.join(accidentes_not_null, how='outer')
```

```python
accidentes_n
```

```python
# merge the two dataframes
accidentes = pd.concat([accidentes_n, accidentes_null])
```

```python
accidentes
```

```python
# order the columns of the dataframe fecha, hora, dia_semana, distrito, localizacion, numero, id, coordenada_x_utm, coordenada_y_utm, longitud, latitud, lesividad, meteo
accidentes = accidentes[['fecha', 'hora', 'dia_semana', 'distrito', 'localizacion', 'numero', 'id', 'coordenada_x_utm', 'coordenada_y_utm', 'longitud', 'latitud', 'lesividad', 'meteo']]
accidentes
```

```python
# order the dataframe by the column 'fecha' and then by the column 'hora'
accidentes = accidentes.sort_values(by=['fecha', 'hora'])
```

```python
# reset the index
accidentes = accidentes.reset_index(drop=True)
```

```python
accidentes
```

```python
# if the last character of the column localizacion is ',', delete it
accidentes['localizacion'] = accidentes['localizacion'].str.rstrip(',')
```
