```python
import pandas as pd
import numpy as np
import time
```

```python
# import data accidentes
accidentes = pd.read_csv('TRAFICO/accidentes.csv')
accidentes
```

```python
accidentes.drop(['Unnamed: 0', 'n_parte'], axis = 1, inplace = True)
```

```python
accidentes
```

```python
#divide datafram into two dataframes by if cooridnates are null or not
accidentes_null = accidentes[accidentes['coordenada_x_utm'].isnull()]
accidentes_not_null = accidentes[accidentes['coordenada_x_utm'].notnull()] 
```

```python
accidentes_null
```

```python
# count hame many rows in localizacion contain 'num'
accidentes_null['localizacion'].str.contains('NUM').sum()
```

```python
accidentes_null['localizacion'].str.contains(' - ').sum()
```

```python
#Count how many are not 0 from the column 'numero'
accidentes_null['numero'].str.contains('0').sum()
```

```python
#Delete NUM from localizacion
accidentes_null['localizacion'] = accidentes_null['localizacion'].str.replace(' NUM', '')
```

```python
#Add the number to the localizacion if it is not 0
accidentes_null['localizacion'] = accidentes_null['localizacion'].str.cat(accidentes_null['numero'], sep = ', ')
```

```python
#Delete ' 0' from localizacion
accidentes_null['localizacion'] = accidentes_null['localizacion'].str.replace(' 0', '')
```

```python
accidentes_null
```

```python
accidentes_null[accidentes_null.localizacion.str.contains('M-30') == True]
```

```python
accidentes_not_null[accidentes_not_null.localizacion.str.contains('M-30') == True]
```

```python
#join the two dataframes
accidentes = pd.concat([accidentes_null, accidentes_not_null])
```

```python
#replace 'CALL.' for 'CALLE' in localizacion
accidentes['localizacion'] = accidentes['localizacion'].str.replace('CALL.', 'CALLE')
```

```python
#save accidentes to csv without index
accidentes.to_csv('TRAFICO/accidentes.csv', index = False)
```
